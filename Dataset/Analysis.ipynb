```python
import plotly.express as px
```

```python
fig = px.density_mapbox(df, lat='lat', lon='long', radius=10,
                        center=dict(lat=0, lon=180), zoom=0,
                        mapbox_style="stamen-terrain")
fig.show()
```

```python
px.set_mapbox_access_token(open("token.mapbox_token").read())
fig = px.scatter_mapbox(df, lat="lat", lon="long", color="disaster_type", 
                  color_continuous_scale=px.colors.cyclical.IceFire)
fig.show()
```

```python
import plotly.figure_factory as ff
fig = ff.create_hexbin_mapbox(
    data_frame=df, lat="lat", lon="long",
    nx_hexagon=200, opacity=0.9, labels={"color": "Point Count"},
    min_count=1,
)
fig.update_layout(margin=dict(b=0, t=0, l=0, r=0))
fig.show()
```
